```python
# A list of number having duplicates value 
# Return the Unique Values

num = [1,2,3,5,2,6,7,1,4,3,5]
print("Original List : ",num)

#print(list(set(num))) -- solution 1

unique = []

for i in num:
    if i not in unique:
        unique.append(i)
print(unique)
```
